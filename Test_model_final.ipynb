Hình 1

```python
import numpy as np
from keras.preprocessing import image
test_image = image.load_img('E:\Code files\ipynb file\Data_test\BanhTrungThu\BanhTrungThu_000303.jpg', target_size = (128, 128))
test_image = image.img_to_array(test_image)

plt.imshow(test_image/255)
test_image = np.expand_dims(test_image, axis = 0)
```

```python
model = keras.models.load_model('model.h5')
```

```python
result = model.predict(test_image)

print(result[0])
```

```python
test_image = image.load_img('E:\Code files\ipynb file\Data_test\BanhTrungThu\BanhTrungThu_000303.jpg', target_size = (128, 128))
test_image = image.img_to_array(test_image)

test_image = test_image / 255
test_image, hog_image = hog(test_image, orientations=9, pixels_per_cell=(8, 8),
                	cells_per_block=(2, 2), visualize=True, multichannel=True)

test_image = np.array(test_image)
np.expand_dims(test_image, axis = 0)
test_image = test_image.reshape(1,-1)
```

Rbf kernel

```python
clf = load('model_clf.joblib')
print(clf.predict(test_image))
```

Poly Kernel

```python
clf_pl = load('model_clf_pl.joblib')
print(clf_pl.predict(test_image))
```
